```python
import matplotlib.pyplot as plt

def piecharts(df:pd.DataFrame, nrows=1, ncols=3) -> None:
    nominal_vars = df.select_dtypes(exclude = 'number').columns
    fig, ax = plt.subplots(nrows, ncols, figsize = (10, 10))
    for i, colname in enumerate(nominal_vars):
        val = df[colname].value_counts()
        ax.flat[i].pie(val.values, labels = val.index, autopct = '%1.1f%%')
        ax.flat[i].set_title(colname)
    plt.tight_layout()
    plt.show()
piecharts(df)
```

```python
def boxplots(df: pd.DataFrame) -> None:
    num_var = df.select_dtypes("number").columns
    plt.figure(figsize=(10, 4))
    for col in enumerate(num_var):
        plt.subplot(1, 3, col[0]+1)
        sns.boxplot(data=df,
                    x="Statut",
                    y=col[1],
                    hue="Genre",
                    gap=.2)
        plt.tight_layout()
    plt.show()

boxplots(df)
```

```python
def countplots(df: pd.DataFrame) -> None:
    cat_var = df.select_dtypes(exclude="number").columns
    plt.figure(figsize=(16, 6))
    for col in enumerate(cat_var):
        plt.subplot(1, 3, col[0]+1)
        sns.countplot(x=col[1], hue='Genre', palette="Paired",data=df, width=.4)
        plt.ylabel("Observations")
    plt.tight_layout()
    plt.legend(loc='center')
    plt.show()
countplots(df)
```

```python
def boxplots(df: pd.DataFrame) -> None:
    num_var = df.select_dtypes("number").columns
    plt.figure(figsize=(10, 4))
    for col in enumerate(num_var):
        plt.subplot(1, 3, col[0]+1)
        sns.boxplot(data=df,
                    x="day",
                    y=col[1],
                    hue="sex",
                    gap=.2)
        plt.tight_layout()
    plt.show()

boxplots(tips)
```

```python
def histograms(df, nrows=1, ncols=3) :
    num_vars = df.select_dtypes(include=['number']).columns
    plt.figure(figsize=(10, 4))
    for col in enumerate(num_vars):
        plt.subplot(nrows, ncols, col[0]+1)
        sns.histplot(x=col[1], data=df, bins=25, kde=True, color="C3")
        plt.ylabel("Nombre d'observations")
    plt.tight_layout()
    plt.savefig("histo.png")
    plt.show()


histograms(tips)
```
